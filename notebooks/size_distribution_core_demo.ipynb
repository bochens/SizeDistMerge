# Size Distribution Core Demo

Small, data-free examples for the public `sizedistmerge` API.

```python
import numpy as np
import sizedistmerge as sdm

mids_nm = np.array([20.0, 40.0, 80.0])
edges_nm = sdm.edges_from_mids_geometric(mids_nm)
dndlog = np.array([1200.0, 800.0, 250.0])

counts = sdm.counts_from_dndlog(dndlog, edges_nm=edges_nm)
target_edges_nm = np.geomspace(edges_nm[0], edges_nm[-1], 6)
rebinned = sdm.remap_dndlog_by_edges_any(edges_nm, target_edges_nm, dndlog)
rebinned_counts = sdm.counts_from_dndlog(rebinned, edges_nm=target_edges_nm)

print("source edges nm:", edges_nm)
print("target edges nm:", target_edges_nm)
print("source total count:", np.nansum(counts))
print("rebinned total count:", np.nansum(rebinned_counts))
```

```python
surface = sdm.dsdlog_from_dndlog(mids_nm, dndlog)
volume = sdm.dvdlog_from_dndlog(mids_nm, dndlog)

print("dS/dlogDp:", surface)
print("dV/dlogDp:", volume)
```

```python
aerodynamic_nm = np.array([500.0, 1000.0, 2000.0])
volume_equivalent_nm = sdm.da_to_dv(aerodynamic_nm, rho_p=1200.0)

print(volume_equivalent_nm)
```
